```python
import math
from scipy import signal
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Get 5 user inputs for all given values

# Passband attenuation
Ap = int(input("Enter Ap in dB: "))

# Stopband attenuation
As = int(input("Enter As in dB: "))

# Passband
wp = float(input("Enter wp: "))

# Stopband
ws = float(input("Enter ws: "))
```

```python
# scipy.signal.buttord(wp, ws, gpass, gstop, analog=False, fs=None)
N, Wn = signal.buttord(wp, ws, Ap, As)
# Output = (lowest_order, Butterworth_natural_frequency)
print("Cutoff frequency [Butterworth] =", Wn)
```

```python
# signal.butter(N, Wn, btype='low', analog=False, output='ba', fs=None)
b, a = signal.butter(N, Wn, 'low')
w, h = signal.freqz(b, a)
plt.semilogx(w, 20 * np.log10(abs(h)))
plt.title('Butterworth filter frequency response')
plt.xlabel('Frequency [Hz]')
plt.ylabel('Amplitude [dB]')
plt.margins(0, 0.1)
plt.grid(which='both', axis='both')
plt.show()
```

```python
# scipy.signal.cheb1ord(wp, ws, gpass, gstop, analog=False, fs=None)
ord, wn = signal.cheb1ord(wp, ws, Ap, As)
# ord = Order; wn = Chebyshev natural frequency
print("Cutoff frequency [Chebyshev] =", wn)
```

```python
# scipy.signal.cheby1(N, rp, Wn, btype='low', analog=False, output='ba', fs=None)
b, a = signal.cheby1(ord, 3, wn)
w, h = signal.freqz(b, a)
plt.semilogx(w, 20 * np.log10(abs(h)))
plt.title('Chebyshev Type I frequency response (rp=3)')
plt.xlabel('Frequency [Hz]')
plt.ylabel('Amplitude [dB]')
plt.margins(0, 0.1)
scale_factor = 10
ymin, ymax = plt.ylim()
plt.ylim(ymin/scale_factor, ymax/scale_factor)
plt.grid(which='both', axis='both')
plt.show()
```
