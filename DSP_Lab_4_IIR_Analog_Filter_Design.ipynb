```python
# Get 5 user inputs for all given values

# Passband attenuation
Ap = int(input("Enter Ap in dB: "))

# Stopband attenuation
As = int(input("Enter As in dB: "))

# Passband edge
fp = int(input("Enter fp in kHz: "))

# Stopband edge
fs = int(input("Enter fs in kHz: "))
```

```python
#Fs
Fs = int(input("Enter Fs in kHz: "))
```

```python
# Find out wp and ws:

wp = (2*math.pi*fp) / Fs

ws = (2*math.pi*fs) / Fs

print("wp =", wp)
print("ws =", ws)
```

```python
# checking for wp in fraction format (pi is rounded off to 3)
from fractions import Fraction

Fraction((2*math.pi*fp) / Fs).limit_denominator(3)

# answer should be pi/3
```

```python
# checking for ws in fraction format (pi is rounded off to 3)
from fractions import Fraction

Fraction((2*math.pi*fs) / Fs).limit_denominator(2)

# answer is pi/2
```

```python
# BUTTERWORTH:

# Passband and Stopband frequencies

omega_s = 2*Fs*math.tan(ws/2)

omega_p = 2*Fs*math.tan(wp/2)

print("Omega_s =", round(omega_s, 3), end=" kHz\n")
print("Omega_p =", round(omega_p, 3), end=" kHz\n")
```

```python
# Calculate minimum value of N

N = (math.log(math.pow((math.pow(10, 0.1*As)-1)/(math.pow(10, 0.1*Ap)-1), 0.5))) / (math.log(omega_s/omega_p))
N = math.ceil(N)
print("N >=", N)
```

```python
# Calculating the cutoff frequency:

omega_cutoff = (omega_p) / (math.pow((math.pow(10, 0.1*Ap)-1), (1/(2*N))))
print("Omega_cutoff =", round(omega_cutoff, 3), end=" kHz\n")
```

```python
# getting cosh inverse:

print(math.cosh(2*math.pi))
import numpy as np

print(np.arccosh(10))
```

```python
# FOR CHEBYSHEV FILTER
# Calculating Order N
N_cheb = np.arccosh(math.pow((math.pow(10, 0.1*As)-1)/(math.pow(10, 0.1*Ap)-1), 0.5)) / np.arccosh(omega_s/omega_p)

N_cheb = math.ceil(N_cheb)
print("N >=", N_cheb)
```

```python
# Calculating Cutoff Frequency:

omega_cutoff_cheb = (omega_p) / (math.pow((math.pow(10, 0.1*Ap)-1), (1/(2*N_cheb))))
print("Omega_cutoff =", round(omega_cutoff_cheb, 3), end=" kHz\n")
```
