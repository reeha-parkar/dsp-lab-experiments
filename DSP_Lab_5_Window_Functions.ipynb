```python
M = int(input("Enter the value of M: "))
```

```python
n = range(0, M)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
#################################### TIME DOMAIN ################################### 
plt1 = plt.figure(figsize=(20, 10))

# Bartlett (triangular)
plt.subplot(2, 2, 1)
bart = []
for i in n:
    yi = 1 - (2*np.abs((i - ((M-1)/2))) / (M-1))
    bart.append(yi)
plt.stem(range(M), bart)
plt.title("Bartlett")
plt.grid(True)

# Blackman
plt.subplot(2, 2, 2)
black = []
for i in n:
    yi = 0.42 - 0.5*(math.cos((2*math.pi*i)/(M-1))) + 0.08*(math.cos((4*math.pi*i)/(M-1)))
    black.append(yi)
plt.stem(range(M), black)
plt.title("Blackman")
plt.grid(True)

# Hamming
plt.subplot(2, 2, 3)
hamm = []
for i in n:
    yi = 0.54 - 0.46*(math.cos((2*math.pi*i)/(M-1)))
    hamm.append(yi)
plt.stem(range(M), hamm)
plt.title("Hamming")
plt.grid(True)

# Hanning
plt.subplot(2, 2, 4)
hann = []
for i in n:
    yi = 0.5*(1 - (math.cos((2*math.pi*i)/(M-1))))
    hann.append(yi)
plt.stem(range(M), hann)
plt.title("Hanning")
plt.grid(True)
plt.show()
```

```python
import scipy
from scipy import signal
```

```python
#################################### FREQUENCY DOMAIN ################################### 

fig = plt.figure(figsize=(20, 15))
ax = fig.subplots(nrows=2, ncols=2)

# Bartlett
w1, h1 = signal.freqz(bart)
ax[0, 0].set_title('Digital filter frequency response for Bartlett')
ax[0, 0].plot(w1, 20 * np.log10(abs(h1)), 'b')
ax[0, 0].set_ylabel('Amplitude [dB]', color='b')
ax[0, 0].set_xlabel('Frequency [rad/sample]')
ax[0, 0].grid(True)

# Blackman
w2, h2 = signal.freqz(black)
ax[0, 1].set_title('Digital filter frequency response for Blackman')
ax[0, 1].plot(w2, 20 * np.log10(abs(h2)), 'b')
ax[0, 1].set_ylabel('Amplitude [dB]', color='b')
ax[0, 1].set_xlabel('Frequency [rad/sample]')
ax[0, 1].grid(True)


# Hamming
w3, h3 = signal.freqz(hamm)
ax[1, 0].set_title('Digital filter frequency response for Hamming')
ax[1, 0].plot(w3, 20 * np.log10(abs(h3)), 'b')
ax[1, 0].set_ylabel('Amplitude [dB]', color='b')
ax[1, 0].set_xlabel('Frequency [rad/sample]')
ax[1, 0].grid(True)


# Hanning
w4, h4 = signal.freqz(hann)
ax[1, 1].set_title('Digital filter frequency response for Hanning')
ax[1, 1].plot(w4, 20 * np.log10(abs(h4)), 'b')
ax[1, 1].set_ylabel('Amplitude [dB]', color='b')
ax[1, 1].set_xlabel('Frequency [rad/sample]')
ax[1, 1].grid(True)

plt.show()
```
